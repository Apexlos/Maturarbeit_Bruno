```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

bild = cv2.imread('/content/IMG_9997.jpg', 1)
print(bild.dtype)
bild = cv2.resize(bild, (0, 0), fx=0.5, fy=0.5)
bild2 = bild[ int(1500 * 0.5) :int(2500 * 0.5) , int(1000 * 0.5):int(4500 * 0.5)]


mask = np.zeros(bild2.shape[:2], np.uint8)#Maske
hinter = np.zeros((1, 65), np.float64)
Vordergrund = np.zeros((1, 65), np.float64)
rectangle = (10, 10, bild2.shape[1] - 10, bild2.shape[0] - 10)
cv2.grabCut(bild2, mask, rectangle, hinter, Vordergrund, 2, cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
image_segmented = bild2 * mask2[:, :, np.newaxis]

plt.imshow(cv2.cvtColor(image_segmented, cv2.COLOR_BGR2RGB))
plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(bild2, cv2.COLOR_BGR2RGB))
plt.show()

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(image_segmented, cv2.COLOR_BGR2RGB))
bilddenoise= cv2.medianBlur(image_segmented,ksize=3)#Blur
denoisergb= cv2.cvtColor(bilddenoise, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15, 5), dpi=300)
linie= cv2.Canny(denoisergb,80,90)#Canny




liniergb= cv2.cvtColor(linie, cv2.COLOR_GRAY2RGB)
plt.imshow(liniergb)
plt.show()


weiss = np.argwhere(linie == 255)#weisse Pixel Finden


weiss_x = weiss[:, 1]
weiss_y = -weiss[:, 0]


hx = max(weiss_x)
minx= min(weiss_x)
hy = max(-weiss_y)
lnge = hx - minx #Lnge berechnen
print(hx,minx,hy)
print(lnge)

plt.show()
```
